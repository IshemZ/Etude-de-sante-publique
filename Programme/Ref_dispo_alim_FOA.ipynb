```python
#Importation de la librairie Pandas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dispo_alim_foa = pd.read_csv('../Source/FAOSTAT_data_fr_2-17-2024.csv')
dispo_alim_foa
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(dispo_alim_foa):
    dispo_alim_foa = dispo_alim_foa.pivot_table(index=['Zone','Produit'], columns='Élément', values='Valeur')
    dispo_alim_foa.reset_index(inplace=True)
    return dispo_alim_foa

dispo_alim_foa_clean = clean_data(dispo_alim_foa.copy())
dispo_alim_foa_clean
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(dispo_alim_foa_clean):
    # Drop column: 'Calories/Année'
    dispo_alim_foa_clean = dispo_alim_foa_clean.drop(columns=['Calories/Année', 'Matière Grasse/Année','Protéines/Année','Population totale'])
    dispo_alim_foa_clean = dispo_alim_foa_clean.rename(columns={'Loss': 'Perte', ''})
    return dispo_alim_foa_clean

dispo_alim_foa_clean_1 = clean_data(dispo_alim_foa_clean.copy())
dispo_alim_foa_clean_1.head()
```
